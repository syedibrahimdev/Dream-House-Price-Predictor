```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.metrics import  mean_squared_error, r2_score
from sklearn.ensemble import RandomForestRegressor
```

```python
df = pd.read_csv(r"E:\ibrahim\Portfolio Projects\House Price Prediction\data\processed\train_processed.csv")
df.head()
```

```python
df['TotalBath'] = df['FullBath'] + 0.5*df['HalfBath'] + df['BsmtFullBath'] + 0.5*df['BsmtHalfBath']
df['TotalPorchSF'] = df['OpenPorchSF'] + df['EnclosedPorch'] + df['3SsnPorch'] + df['ScreenPorch'] + df['WoodDeckSF']
df['HouseAge'] = df['YrSold'].astype(int) - df['YearBuilt'].astype(int)
df['RemodAge'] = df['YrSold'].astype(int) - df['YearRemodAdd'].astype(int)
```

```python
x = df.drop('SalePrice', axis=1)
y = df['SalePrice']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
lr = LinearRegression()
lr.fit(x_train, y_train)

y_pred = lr.predict(x_test)
```

```python
ridge = Ridge(alpha=1.0)
ridge.fit(x_train, y_train)
y_ridge = ridge.predict(x_test)

lasso = Lasso(alpha=0.001)
lasso.fit(x_train, y_train)
y_lasso = lasso.predict(x_test)
```

```python
rf = RandomForestRegressor(n_estimators=500, random_state=42)
rf.fit(x_train, y_train)
y_rf = rf.predict(x_test)
```

```python
def evaluate(y_true, y_pred, name="Model"):
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    r2 = r2_score(y_true, y_pred)
    print(f"{name} -> RMSE: {rmse:.4f}, R2: {r2:.4f}")

evaluate(y_test, y_pred, "Linear Regression")
evaluate(y_test, y_ridge, "Ridge Regression")
evaluate(y_test, y_lasso, "Lasso Regression")
evaluate(y_test, y_rf, "Random Forest")
```

```python
y_pred_actual = np.expm1(y_rf)
```

```python
feature_importance = pd.Series(rf.feature_importances_, index=x_train.columns)
top_features = feature_importance.sort_values(ascending=False).head(20)

plt.figure(figsize=(10,6))
sns.barplot(x=top_features, y=top_features.index)
plt.title("Top 20 Feature Importances")
plt.show()
```
